## Start CAS session

```sas
option casport=5570 cashost="localhost";
cas casauto;
caslib _all_ assign;
```

```sas
proc datasets library=public memtype=data;
run;
```

```sas
proc logselect data=Public.readmission_analysis;
    class readmit_number race_cd discharged_to dx_group;
    model readmit_number = race_cd discharged_to dx_group  order_total_charges admit_month 
    order_set_used operationcount op_visits6 PatientAge ;
    selection method=backward;
    partition rolevar = _partind_(train='0' valid='1' test='2');
run;
```
